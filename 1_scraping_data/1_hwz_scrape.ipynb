```python
#currently all posts of every page - text and time 
posts = []

for link in links:
    page = 1

    while True:

        print(f'scraping thread - page {page}')

        res = requests.get(f'{link}-{str(page)}.html')
        soup = BeautifulSoup(res.text, 'html.parser')
        post_div = soup.find_all('div', {'class': 'post-wrapper'})

        for post in post_div:
            posts.append(post.find('div',{'class': 'post_message'}).text.replace('\r', '').replace('\n', '').replace('\t', ''))
            posts.append(post.find('td', {'class':'thead'}).text.replace('\r', '').replace('\n', '').replace('\t', ''))
            posts.append(soup.find('h2', {'class': 'header-gray'}).text.replace('\r', '').replace('\n', '').replace('\t', ''))
        sleep(randint(1,4))

        try:   
            next_page = soup.find('span', {'class': 'desc'}).text.split("of",1)[1].strip()
            if next_page != str(page):
                page +=1
            else:
                print('all pages in thread scraped')
                break
        except AttributeError:
            print('thread only has one page')
            break
```
