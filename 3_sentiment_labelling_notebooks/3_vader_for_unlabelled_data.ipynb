```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
#import train data 
posts = pd.read_csv('./scrape/train_posts.csv')

display(posts.shape)
display(posts.head())

posts.fillna('nopost', inplace=True)
display(posts.isnull().sum())
```

```python
#taking sample of 500 posts preprocessed post_clean (ref to data clean and EDA notebook)
sample_posts = posts['post_clean'][0:500]

display(sample_posts.shape)
sample_posts.head()
```

```python
#sentiment analyser with VADER

sid = SentimentIntensityAnalyzer()
```

```python
sid.polarity_scores('Humans not great')

#use compound score
#metric that calculates the sum of all the lexicon ratings which have been normalized between -1(most extreme negative) and +1 (most extreme positive)
```

```python
vader_sentiment = []

for i in range(len(sample_posts)):
    vader_sentiment.append(sid.polarity_scores(sample_posts[i])['pos'])
```

```python
sample_vader_sentiment = pd.DataFrame(list(zip(sample_posts, vader_sentiment)), 
                                     columns = ['posts', 'vader_sentiment'])

sample_vader_sentiment.head(20)
```

```python
sample_vader_sentiment.to_csv('./vader_sentiment_sample_1.csv')
```
