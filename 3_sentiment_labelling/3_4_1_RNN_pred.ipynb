## Using Recurrent Neural Networks to label sentiment on unlabelled posts 

Having manually labelled 1000 posts from the train set, these labelled posts were used to train Recurrent Neural Networks to incrementally label the remaining 7000 posts in the train set.

Process:
1. Manually label 1000 posts 
2. Split posts to train and val set
3. Train models
4. Evaluate best performing model on val set 
5. Use models to predict next 1000 posts 
6. Manually check accuracy of predictions, make changes to labels 
7. Combine all labelled posts 
8. Repeat 2-7 until all posts in train set labelled

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from imblearn.over_sampling import SMOTE, RandomOverSampler

from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense, Dropout, GRU, Embedding, SpatialDropout1D, LSTM
from keras.callbacks import EarlyStopping
from keras.optimizers import Adam

from sklearn import metrics
```

```python
sample_posts = pd.read_csv('./train_2000_labelled.csv')
sample_posts.head()
```

```python
sample_posts['manual label'].fillna(0, inplace=True)
```

```python
sample_posts.isnull().sum()
```

```python
sample_posts.fillna('nopost', inplace=True)
```

```python
# function for preprocessing post in sample_post 
# will not remove stop words to help with sequentiality 

def preprocess(word):
     
    # tokenize and convert lower 
    # \w also removes punctuation - may need to add extra no punc if tokenizing does not do it
    token = RegexpTokenizer(r'\w+')
    tokens = token.tokenize(word.lower())
 
    #lemmatize words
    lemmatizer = WordNetLemmatizer()
    lem = [lemmatizer.lemmatize(word) for word in tokens]
    
    #return words as a single string 
    return(' '.join(lem))
```

```python
post_clean_for_rnn = []

for p in sample_posts["post"]:
    post_clean_for_rnn.append(preprocess(p))

print(f"checking post_clean_for_rnn: \n{post_clean_for_rnn[1:3]}")

sample_posts['post_clean_for_rnn'] = post_clean_for_rnn
sample_posts.head(10)
```

```python
#classes v unbalanced
ax = sns.countplot(sample_posts['manual label'])
```

```python
#Vectorize post text, by turning each text into either a sequence of integers or into a vector

# The maximum number of words to be used. (most frequent)
max_nb_words = 50000

# Max number of words in each complaint.
max_seq_length = 512

# This is fixed.
embedding_dim = 100

tokenizer = Tokenizer(num_words=max_nb_words, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(sample_posts['post_clean_for_rnn'].values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
#truncate and pad 
X = tokenizer.texts_to_sequences(sample_posts['post_clean_for_rnn'].values)
X = pad_sequences(X, maxlen=max_seq_length)
print('Shape of data tensor:', X.shape)
```

```python
#one hot encoding labels 
y = pd.get_dummies(sample_posts['manual label']).values
y.shape
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = .2, 
                                                 stratify = y, random_state = 42)
```

```python
print(X_train.shape,y_train.shape)
print(X_val.shape,y_val.shape)
```

```python
#SMOTE 

sm = SMOTE()
X_train, y_train = (X_train, y_train)
```

```python
# Scale
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_val = ss.transform(X_val)
```

```python
model = Sequential()
model.add(Embedding(max_nb_words, embedding_dim, input_length=X.shape[1]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(100, dropout=0.5, recurrent_dropout=0.5))
# model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))

opt = Adam(learning_rate=0.001)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['AUC'])

model.summary()
```

```python
epochs = 10
batch_size = 32

# class_weight = {0: .8,
#                 1: .5,
#                2: 1}

history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size,
                    validation_data = (X_val, y_val),
#                     callbacks=[EarlyStopping(monitor='val_loss', patience=5, min_delta=0.0001)],
#                     class_weight = class_weight,
                   verbose=1)
```

```python
acc = model.evaluate(X_val,y_val)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(acc[0],acc[1]))
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend();
```

```python
#fit X and y to model to predict other unlabelled posts 

model.fit(X, y, epochs=10, batch_size=batch_size, verbose=1)
```

```python
#import unlabelled data to pred sentiment 

unlabelled = pd.read_csv('./unlabelled_posts/2000_2999_unlabelled.csv')
unlabelled.head()
```

```python
unlabelled_posts = unlabelled['post']
```

```python
#predict sentiment class of post in unlabelled set (2000-2999)
posts_to_pred = []

for post in unlabelled_posts:
    seq = tokenizer.texts_to_sequences(post)
    padded = pad_sequences(seq, maxlen=max_seq_length)
    posts_to_pred.append(padded)
    
```

```python
len(posts_to_pred)
```

```python
posts_to_pred[1]
```

```python
preds = []

for i in range(len(posts_to_pred)):
    pred = model.predict_classes(posts_to_pred[i], verbose = 1)
    preds.append(pred.sum()/len(pred))
```

```python
print(len(preds))
preds[1]
```

```python
unlabelled['predicted_labels'] = preds 
unlabelled.tail()
```
